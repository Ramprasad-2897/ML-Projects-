# WEB SCRAPPING

```python
#IMPORT PACKAGES

import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
#Pointing to URL
url = "https://timesofindia.indiatimes.com/home/education/news/career-in-data-science-in-india-courses-certification-salary-future-growth/articleshow/75023624.cms"
```

```python
#sending the request to url, fetching the content of url

response=requests.get(url)
```

```python
response      # if the output is 200 then the request went well, else there is some problem
```

```python
#initialising parser
soup = BeautifulSoup(response.text,"html.parser")
```

```python
soup        # soup contains all the html content from the webpage. But we dont want everything. Only the article.
```

```python
#check what we need on  web page
```

```python
#Fetching Div = " section 1" data

content = soup.find('div',{"class":"section1"})
```

```python
content # but content will contain all the  text is there but also the structure of the html page
```

```python
# we need the plain text
```

```python
#creating the text data

article = ''
for i in content.findAll('div'):
    article = article + '' + i.text

article
```

```python
print(article)
```

```python
#NLP
import nltk
#importing the stop words
from nltk.corpus import stopwords
#importing the lemmatizer
from nltk.stem import WordNetLemmatizer
#importing the regular expression
import re
from nltk.stem import PorterStemmer
#importing portstemmer for stemming the data
ps=PorterStemmer()
#tokenizer for the splitting the data into tokens
articl= nltk.sent_tokenize(article)
ld=WordNetLemmatizer()
```

```python
articl
sentence=[]
```

```python
for i in range(len(articl)):
    
    artil=re.sub('[^a-zA-Z]',' ',articl[i])
    artil=re.sub(r'#','',articl[i])
    #converting to lower case
    artil=artil.lower()
    #splitting the paragraph into sentences
    artil=artil.split()
    #lemmatizing the words where we get the meaning ful words
    artil=[ld.lemmatize(word) for word in  artil if not word in set(stopwords.words('english'))]
    #joining the sentences back after splitting the data
    artil=' '.join(artil)
    #appending the data inserting to the list
    sentence.append(artil)
    #printing the data
    print(sentence)
    
```

```python
#importing the pandas for making dataframe of the data
import pandas as pd
#converting the paragraph data to the dataframes
b=pd.DataFrame(sentence)
b
```

```python
from textblob import TextBlob
```

```python
#creating the function for checking the polarity values(emotions in the data)
def getTextPolarity(txt):

    return TextBlob(txt).sentiment.polarity # emotions in particular  positive, negative, neutral
```

```python
#calling the polarity function
b['pol']=b[0].apply(getTextPolarity)
```

```python
b
```

```python
#function for the labeling the negative positive and neutral data

def getanalysis(a):
    if a<0:
        return "Negative"
    elif a==0:
        return "neutral"
    else:
        return "positive"
```

```python
#calling the analysis function
b['score']=b['pol'].apply(getanalysis)
```

```python
#printing the data
b
```

```python
#positive percentage
positive=b[b['score']=='positive']
print(str(positive.shape[0]/(b.shape[0])*100)+"% of postive data")

pos=positive.shape[0]/b.shape[0]*100
```

```python
#negative percentage
neuta=b[b['score']=='neutral']
print(str(neuta.shape[0]/(b.shape[0])*100)+"% of neutral data")
neuta=neuta.shape[0]/b.shape[0]*100
```

```python
#neutral percentage
neg=b[b['score']=='Negative']
print(str(neg.shape[0]/(b.shape[0])*100)+"% of negative data")
neg=neg.shape[0]/b.shape[0]*100
```

```python
#visualizing the data in percentages of polarity(positive, negative and neutral)
import matplotlib.pyplot as plt
explode=(0,0.1,0)
labels='positive','neutral','negative'
sizes=(pos,neg,neuta)
colors=['yellowgreen','lightcoral','gold']
plt.pie(sizes,explode=explode,colors=colors,autopct='%1.1f%%',startangle=120)
plt.legend(labels,loc=(-0.5,0.5),shadow=True)
plt.axis('equal')
```
